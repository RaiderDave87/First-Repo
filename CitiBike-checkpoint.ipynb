```python
import pandas as pd
```

```python
first_data = pd.read_csv("201811-citibike-tripdata.csv")
combined_data = pd.DataFrame(first_data)
combined_data.head()
```

```python
for i in range(201804, 201812):
    file_to_load = str(i) + "-citibike-tripdata.csv"
```

```python
len(combined_data.index)
```

```python
starttime2 = pd.to_datetime(combined_data.starttime, errors='coerce')
stoptime2 = pd.to_datetime(combined_data.stoptime, errors='coerce')
new_data = combined_data.assign(start_date=starttime2.dt.date, start_hour=starttime2.dt.hour,
                                stop_date = stoptime2.dt.date, stop_hour=stoptime2.dt.hour)
new_data.replace({'gender': {0: 'Unknown', 1: 'Male', 2: 'Female'}}, inplace=True)
```
